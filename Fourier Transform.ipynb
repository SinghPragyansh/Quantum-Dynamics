```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as sci
import scipy as sc
import scipy.fftpack as sft
```

```python
#Creating X GRID
xmin = 0
xmax = 10
dx = 1
x = np.arange(xmin,xmax,dx)
print(x)        #We have created a list for even number of x,therefore spatial freq components will follow the needed formula.
```

```python
# Creating spatial frequency and then subsequently K GRID
N = len(x)
xi = sft.fftfreq(N,dx)

#AS WE CAN SEE, fftfreq ALREADY CONTAINS ALGO FOR SPATIAL FREQ COMPUTATION, NOW WE JUST NEED TO CALCULATE K GRID

k = 2*np.pi*xi
print(k)          # Obtained k list also follows discrete f.t. result,where dk*dx = 2*pi/N 
```

```python
# NEW X Grid of odd number of x
xmin = 0
xmax = 11
dx = 0.1
x = np.arange(xmin,xmax,dx)
N = len(x)
xi = sft.fftfreq(N,dx)

k=2*np.pi*xi
```

```python
# Define the Cosine function in position or real space
y = np.cos(20*x)
plt.plot(x,y)
```

```python
#Now do the FFT for k domain (Momentum or Reciprocal)

y_k = sft.fft(y)
plt.plot(k,abs(y_k))     #Using "abs" , for taking in only real part of the transformed function 
```
