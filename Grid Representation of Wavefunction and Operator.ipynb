```python
# We will focus on wavefunction representation for programming
# In the rep, we express the continuos w.f. on a set of position grid points, like x min and x max, which will be boundary condition 
# hence, w.f. will be zero at the extreme points and the problem domain(position space) will consist of discrete grid points.
#Therefore, we take discretised w.f. in between the boundary points.

import scipy as sc
import scipy.integrate as nmi
import numpy as np
import matplotlib.pyplot as plt
```

```python
xmin= -100
xmax = 100.1
dx=0.1
x = np.arange(xmin,xmax,dx)
l = len(x)
print(l)
print(x)

```

```python
# Giving input for a gaussian function in position space
psi = np.exp(-(x-7)**2)
plt.plot(x,psi,'*-')
```

```python
some_den =[]
ProbDen = abs(psi)**2
plt.plot(x,ProbDen)
for i in ProbDen:         #Checking the points on grid where the prob is not zero
    if i != 0.0:
        some_den.append(i)    #making a array out of non-zero prob density values
print(ProbDen)            #Prob Density is zero on most of the points of the grid
print(some_den)
```

```python
# For NORMALISATION, we will make use of numerical method integration
```

```python
Norm = np.sqrt(nmi.simps(ProbDen,x))
print(Norm)
N_psi = psi/Norm
```

```python
Norm_PD = abs(N_psi)**2
print(Norm_PD)
plt.plot(x,Norm_PD,color="orange")
```

```python
# EXPECTATION VALUE OF POSITION

Integrand = Norm_PD*x
Exp_X = nmi.simps(Integrand,x)
print(Exp_X)

# OBTAINED EXPECTATION POSITION IS THE PEAK OR AVERAGE OF THIS GAUSSIAN FUNCTION, AS VISIBLE IN FUNCTION AS WELL.
```
