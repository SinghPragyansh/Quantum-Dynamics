```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import scipy.fftpack as sft
```

```python
xmin = -100
xmax = 100
dx = 0.001
x = np.arange(xmin,xmax,dx)
y = np.exp(-x**2)*np.cos(20*x)   #CREATING TRAVELLING GAUSSIAN WAVEPACKET
plt.plot(x,y)
plt.xlim(-5,5)

```

```python


N = len(x)
xi = sft.fftfreq(N,dx)
k = 2*np.pi*xi

y_k = sft.fft(y)         
plt.plot(k,abs(y_k))
plt.xlim(-30,30)



```
